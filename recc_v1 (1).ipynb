```python
# date / time, probability 

def recommend(hour, origin, destination, threshold):
    
    a = []

    max_hour = hour + threshold
    min_hour = hour - threshold
    
    for x in np.arange(min_hour, max_hour + 1):
        prediction = predict_delay(str(x), origin, destination)
        probability = 1 - prediction
        a.append(probability)
    
    recommendations = pd.DataFrame({"Time": list(np.arange(min_hour, max_hour + 1)),
                                   "Probability of Delay": a})
        
    return recommendations
```

```python
recommend(11, "JFK", "ATL", 2)
```
